### BUILDING THE MACHINE LEARNING ALGORITHM

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
```

```python
# Splitting the dataset to train and test data
X = ccfraud.iloc[:,:-1]
Y = ccfraud['Class']
# Where Y represents the target variable and X represents the predictor variable
```

```python
X
```

```python
Y
```

```python
X_train, x_test, Y_train, y_test = train_test_split(X, Y, test_size = 0.30, random_state = 410)
```

```python
#Creating an instance of a Logistic regression named lr
help(LogisticRegression())
```

```python
lr = linear_model.LogisticRegression(C=1e5)
```

```python
lr.fit(X_train, Y_train)
```

```python
Y_pred = np.array(lr.predict(x_test))
y = np.array(y_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
print(confusion_matrix(y_test, Y_pred))
```

```python
print(accuracy_score(y_test, Y_pred))
```

```python
print(classification_report(y_test, Y_pred))
```
